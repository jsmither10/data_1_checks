```python
import requests
import pandas as pd
import json
```

```python
#accessing the API
r = requests.get('https://peoplegeneratorapi.live/api/person/100')
print(r.status_code)
print(r.json())
```

```python
#viewing the data as a dataframe
data=r.json()
df = (pd.DataFrame(data))
df.head()
```

```python
#getting a list of all the column names
column_names = df.columns.tolist()
print(column_names)
```

```python
#DESCRIPTIVE STAT NUMBER 1 - AGE
min_age = df["age"].min()
print("The youngest person on the list is " + str(min_age))

max_age = df["age"].max()
print("The oldest person on the list is " + str(max_age))

avg_age = df["age"].mean()
print("The average age of the person on the list is " + str(avg_age))
```

```python
#DESCRIPTIVE STAT NUMBER 2 - Has Children

has_children = df["hasChildren"].value_counts()
print(has_children.to_string())

has_children_true_count = (df["hasChildren"] == True).sum()
print("Out of the 100 people " + str(has_children_true_count) + " have children")

```

```python
#Write a query in Pandas to select a particular set of data. This is a mask of users who have children
mask = df["hasChildren"] == True
maskdf = df[mask]
print(maskdf)
```

```python
#Select and print the SECOND AND THIRD columns of your data frame.

columns_print = [1,2]
selected_columns = maskdf.iloc[:, columns_print]
print(selected_columns)
```
