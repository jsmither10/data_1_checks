```python
import pandas as pd
```

```python
df = pd.read_csv('assets\movies.csv')
df.head()
```

```python
#finding null values
df.isnull().sum().sort_values(ascending=False)
```

```python
#replacing null values in Rating column
df["RATING"] = df["RATING"].fillna(0)
```

```python
#checking our work
df.isnull().sum().sort_values(ascending=False)
```

```python
#removing the rows that include a Null Genre
df.dropna(subset=['GENRE'])
```
